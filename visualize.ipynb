```python
import pandas as pd
import numpy as np
import os
from glob import glob
import random
import matplotlib.pylab as plt
%matplotlib inline
import cv2
import random
```

```python
data_df = pd.read_csv('../outputs/target_idx.csv')
data_df.head()
```

```python
target_idx_list = data_df[data_df['Emphysema'] == 1]['Image Index'].tolist()
control_idx_list = data_df[data_df['Emphysema'] == 0]['Image Index'].tolist()
```

```python
base_path = images_path = '../inputs/images/'

def show_img(idx_list, samples=10, title='No Finding'):
    random.seed(1)
    random.shuffle(idx_list)
    plt.figure(figsize=(14, 6))
    for i in range(samples):
        img_path = base_path + idx_list[i]
        print(img_path)
        img = cv2.imread(img_path)
        plt.subplot(2, 5, i + 1)
        plt.title(title)
        plt.axis("off")
        plt.imshow(img)
    plt.show()
```

### show images of target group

```python
show_img(target_idx_list, title='Emphysema')
```

### show images of control group

```python
show_img(control_idx_list, title='No Finding')
```
